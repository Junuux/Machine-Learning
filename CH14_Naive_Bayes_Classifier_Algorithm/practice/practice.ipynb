<h4>1. 데이터 불러오기</h4>

```python
from sklearn import datasets

wine = datasets.load_wine()
```

```python
dir(wine)
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
import pandas as pd

df = pd.DataFrame(wine.data, columns= wine.feature_names)
df.head()
```

```python
df['target'] = wine.target
df
```

```python
# df_pred =df.sample(n=20, replace=False)
# df.describe()
```

<h4>2. train / test data</h4>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis='columns'), df.target, test_size=0.3, random_state=1)
# X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3, random_state=1
```

```python
y_train.unique()
```

<h4>3. Model</h4>

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB

model_G = GaussianNB()
model_G.fit(X_train, y_train)
print('GaussianNB :', model_G.score(X_test, y_test))

model_M = MultinomialNB()
model_M.fit(X_train, y_train)
print('MultinomialNB :', model_M.score(X_test, y_test))
```
